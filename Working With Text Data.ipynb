```python
sr = pd.Series([10, 25, 3, 11, 24, 6])
 
# Create the Index
index_ = ['Coca Cola', 'Sprite', 'Coke', 'Fanta', 'Dew', 'ThumbsUp']
 
# set the index
sr.index = index_
sr.replace(to_replace = 3, value=200)
```

```python
sr = pd.Series(['New York', 'Chicago', 'Toronto', 'Lisbon', 'Rio'])
 
# Create the Index
index_ = ['City 1', 'City 2', 'City 3', 'City 4', 'City 5'] 
 
# set the index
sr.index = index_
sr.replace(to_replace = ['New York', 'Rio'], value = ['London', 'Brisbane'])
```
